#### Since we cannot perform mean, median and standard deviation for time data.
#### Converting the time data into secods(numerical) format for calcualtion


```python
#Creating Function to convert time(HH:MM:SS) into seconds
def time_to_seconds(time_str):
    if pd.isna(time_str):  # If provided time data is null/NaN values leave as it is
        return np.nan
    hh, mm, ss = map(int, time_str.split(':'))
    return hh * 3600 + mm * 60 + ss


from datetime import timedelta

#Creating Function to convert seconds into time (HH:MM:SS) format

def seconds_to_time(seconds_value):
    td = timedelta(seconds=seconds_value)
    return td
```

```python
#Using LOC for keeping the newly created column permanently with the calls_df dataframe

calls_df.loc[:,'turn_around_time_in_sec'] = calls_df['turn_around_time'].apply(time_to_seconds)
calls_df.loc[:,'served_time_in_sec'] = calls_df['served_time'].apply(time_to_seconds)
calls_df.loc[:,'waiting_time_in_sec'] = calls_df['waiting_time'].apply(time_to_seconds)

calls_df.to_csv('CSV Files/calls.csv', index=False)
```

# Mean

```python
waiting_time_mean = np.round(calls_df['waiting_time_in_sec'].mean())
served_time_mean = np.round(calls_df['served_time_in_sec'].mean())
turn_around_time_mean = np.round(calls_df['turn_around_time_in_sec'].mean())

print(f"Waiting Time Mean(sec): {waiting_time_mean}")
print(f"Served Time Mean(sec) {served_time_mean}")
print(f"Turn Around Time Mean (sec) {turn_around_time_mean}")

```

```python
waiting_mean_time = seconds_to_time(waiting_time_mean)
served_mean_time = seconds_to_time(served_time_mean)
turn_around_mean_time = seconds_to_time(served_time_mean)


print(f"Waiting Time Mean(Time): {waiting_mean_time}")
print(f"Served Time Mean(Time) {served_mean_time}")
print(f"Turn Around Time Mean (Time) {turn_around_mean_time}")
```

# Median

```python
waiting_time_median = np.round(calls_df['waiting_time_in_sec'].median())
served_time_median = np.round(calls_df['served_time_in_sec'].median())
turn_around_time_median = np.round(calls_df['turn_around_time_in_sec'].median())


print(f"Waiting Time Median(sec): {waiting_time_median}")
print(f"Served Time Median(sec): {waiting_time_median}")
print(f"Turn Around Time Median (sec): {waiting_time_median}")
```

```python
waiting_median_time = seconds_to_time(waiting_time_median)
served_median_time = seconds_to_time(served_time_median)
turn_around_median_time = seconds_to_time(turn_around_time_median)


print(f"Waiting Time Mean(Time): {waiting_median_time}")
print(f"Served Time Mean(Time) {served_median_time}")
print(f"Turn Around Time Mean (Time) {turn_around_median_time}")
```

# Standard Deviation

```python
waiting_time_sd = np.round(calls_df['waiting_time_in_sec'].std())
served_time_sd = np.round(calls_df['served_time_in_sec'].std())
turn_around_time_sd = np.round(calls_df['turn_around_time_in_sec'].std())


print(f"Waiting Time Standard Deviation(sec): {waiting_time_sd}")
print(f"Served Time Standard Deviation(sec): {served_time_sd}")
print(f"Turn Around Time Standard Deviation (sec): {turn_around_time_sd}")
```

```python
waiting_sd_time = seconds_to_time(waiting_time_sd)
served_sd_time = seconds_to_time(served_time_sd)
turn_around_sd_time = seconds_to_time(turn_around_time_sd)


print(f"Waiting Time Mean(Time): {waiting_sd_time}")
print(f"Served Time Mean(Time) {served_sd_time}")
print(f"Turn Around Time Mean (Time) {turn_around_sd_time}")
```
