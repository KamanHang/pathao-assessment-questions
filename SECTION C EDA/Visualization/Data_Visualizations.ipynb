# Number of Customers Served per day

```python
def getDayFunction(dateData):
    # Ensure dateData is converted to datetime
    started_date = pd.to_datetime(dateData)

    # If dateData is a Series, use vectorized operations
    if isinstance(started_date, pd.Series):
        day_name = started_date.dt.strftime("%A")
    else:  # If it's a single value, use the standard datetime method
        day_name = started_date.strftime("%A")

    return day_name

calls_df.loc[:,'Day'] = calls_df['started_date'].apply(getDayFunction)
```
